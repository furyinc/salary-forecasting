```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load the dataset
file_path = r"C:\Users\Frila\PycharmProjects\data_salaries\Resources\sdmx_data_500.csv"
data = pd.read_csv(file_path)

# Extract year columns (make sure these match your CSV header names)
years = ['2017', '2018', '2019', '2020', '2021', '2022', '2023']

# Select the row for the Republic of Uzbekistan (first row in this dataset)
region_data = data.loc[0, years].astype(float)

# Convert the series to a DataFrame for time series analysis
df = pd.DataFrame({
    'Year': pd.to_numeric(years),
    'Salary': region_data.values
})

# Prepare data for modeling
X = df['Year'].values.reshape(-1, 1)  # Predictor: Year
y = df['Salary'].values               # Response: Salary

# Initialize and fit the Linear Regression model
model = LinearRegression()
model.fit(X, y)

# Forecast for the next year, e.g., 2024
next_year = np.array([[2024]])
predicted_salary = model.predict(next_year)
print(f"Predicted Average Monthly Salary for 2024: {predicted_salary[0]:.2f} UZS")



# Extend the data for visualization
X_extended = np.append(X.flatten(), 2024)
y_extended = np.append(y, predicted_salary)

plt.figure(figsize=(10, 6))
plt.plot(X, y, marker='o', label='Historical Salary')
plt.plot(2024, predicted_salary, 'ro', markersize=10, label='Forecasted 2024 Salary')
plt.title('Salary Trend and 2024 Forecast (Republic of Uzbekistan)')
plt.xlabel('Year')
plt.ylabel('Salary (UZS)')
plt.grid(True)
plt.legend()
plt.show()
```
